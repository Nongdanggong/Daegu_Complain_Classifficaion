```python
def csv_to_parquet(df, save_name):
    df.to_parquet(f'./{save_name}.parquet')
    del df
    gc.collect()
    print(save_name, 'Done.')
```

```python
# 정규표현식 사용, 불용어 제거
import re
def get_only_hangul(line):
	parseText= ''.join(re.compile('[ |가-힣]*').findall(line))

	return parseText
```

```python
for i, v in enumerate(name_list):
    complain_list_ori[i]['content'] = complain_list_ori[i]['content'].apply(lambda x : get_only_hangul(x))
```
